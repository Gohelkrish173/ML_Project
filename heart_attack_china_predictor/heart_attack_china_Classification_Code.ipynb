```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler , LabelEncoder
```

```python
dt = pd.read_csv("heart_attack_china.csv")

pd.set_option('display.max_columns',None)
dt
```

```python
categorical_cols = dt.select_dtypes(include=['object']).columns
categorical_cols
```

```python
dt.shape
```

```python
x = dt.iloc[:,1:27]
y = dt.iloc[:,27:]
```

```python
LE = LabelEncoder()
```

```python
DT = pd.DataFrame()
DT['Age'] = dt['Age']
DT['Blood_Pressure'] = dt['Blood_Pressure']
DT['CVD_Risk_Score'] = dt['CVD_Risk_Score']
DT
```

```python
mappings = {}

for col in categorical_cols:
    DT[col] = LE.fit_transform(dt[col])  # Convert to numeric
    mappings[col] = dict(zip(LE.classes_, LE.transform(LE.classes_)))

DT

#for col in dt.iloc[:,1:28]:
#    DT[col] = LE.fit_transform(dt[col])

#DT
```

```python
DT.to_csv("heart_attack_china_Classified.csv",index=False)
```

```python
with open("output.txt", "w") as file:
    for key, value in mappings.items():
        file.write(f"{key}: {value}\n")
```
